```python
from sklearn.preprocessing import MinMaxScaler
```

```python
numeric_var = ['Age', 'Total Spend', 'Items Purchased', 'Average Rating', 'Days Since Last Purchase']
numeric_df = df[numeric_var]
```

```python
minmax = MinMaxScaler()
```

```python
scaled_df = minmax.fit_transform(numeric_df)
```

```python
df[numeric_var] = scaled_df
```

```python
df.head()
```

```python
sns.heatmap(df.select_dtypes(exclude='object').corr())
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=3, random_state=41)
```

```python
kmeans.fit(df.drop(['Satisfaction Level', 'Customer ID'], axis=1))
```

```python
kmeans.cluster_centers_
```

```python
kmeans.labels_
```

```python
X = df.drop(['Satisfaction Level', 'Customer ID'], axis=1)
plt.scatter(X.iloc[:, 0], X.iloc[:, 1], c=kmeans.labels_, cmap='Set1', alpha=0.3)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], marker='x', s=100, c='blue')  # Plot cluster centers
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('K-Means Clustering with Cluster Centers')
plt.show()
```

```python
df.head()
```

```python
def sat_levels(x):
    if x == 'Satisfied':
        return 1
    elif x == 'Neutral':
        return 2
    else:
        return 0
```

```python
df['Cluster'] = df['Satisfaction Level'].apply(sat_levels)
```

```python
df.head()
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print(classification_report(df['Cluster'], kmeans.labels_))
print(confusion_matrix(df['Cluster'], kmeans.labels_))
```

## 98% Accuracy Obtained by KMeans Clustering
